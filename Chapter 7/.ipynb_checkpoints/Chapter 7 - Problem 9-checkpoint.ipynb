**9.** This question uses the variables dis (the weighted mean of distances
to five Boston employment centers) and nox (nitrogen oxides concentration
in parts per 10 million) from the Boston data. We will treat
dis as the predictor and nox as the response.

**A.** Use the poly() function to fit a cubic polynomial regression to
predict nox using dis. Report the regression output, and plot
the resulting data and polynomial fits.

```r
library(MASS)
library(boot)
require(splines)
```

```r
poly.fit <- lm(nox~poly(dis,3),data = Boston)
summary(poly.fit)
```

```r
dis.range <- seq(from = min(Boston$dis),to = max(Boston$dis), length = 100)
dis.pred <- predict(poly.fit, newdata = list(dis = dis.range),se = TRUE)
se.bands <- cbind(dis.pred$fit + 2*dis.pred$se,dis.pred$fit - 2*dis.pred$se)
plot(Boston$dis,Boston$nox,main = "Distance to employment centers vs NOX levels",
    ylab = "NOX levels", xlab = "Weighted mean of distances to five Boston employment centers")
lines(dis.range,dis.pred$fit, col = "blue")
matlines(dis.range, se.bands, col = "green",lwd = 2, lty = 2)
```

The regression shows that all terms in the polynomial are significant. The plot exhibits a good fit of the polynomial to the data however it also shows a tail problem with the polynomial fit at higher distances.

**B.** Plot the polynomial fits for a range of different polynomial
degrees (say, from 1 to 10), and report the associated residual
sum of squares.

```r
degree <- 1:10
RSS <- rep(NA,10)

for (d in degree){
    poly.fit <- lm(nox~poly(dis,d),data = Boston)
    distance.pred <- predict(poly.fit,newdata = Boston)
    RSS[d] <- mean((Boston$nox - distance.pred)^2)
}

plot(RSS,col = "red", lwd = 1 ,type = "b")
```

**C.** Perform cross-validation or another approach to select the optimal
degree for the polynomial, and explain your results.

```r
cv.error <- rep(NA,11)

for (d in 1:10){
    poly.fit <- glm(nox~poly(dis,d),data = Boston)
    cross.val <- cv.glm(poly.fit,data = Boston, K = 11)
    cv.error[d] <- cross.val$delta[1] 
}

cv.error
plot(cv.error, main = "11-fold cross validation", xlab = "Degree of polynomial", ylab = "Cross Validation Error", type = "b", col = "Blue")
```

```r
degree <- which.min(cv.error)
print(degree)
```

The lowest cross validation error is obtained at 3-degree polynomial, which is a low-order model and much simpler model when compared to high order polynomials. Therefore this is the optimal degree. 

**D.** Use the bs() function to fit a regression spline to predict nox
using dis. Report the output for the fit using four degrees of
freedom. How did you choose the knots? Plot the resulting fit.

```r
spline.fit <- lm(nox~ bs(dis,df = 4),data = Boston)
dis.pred.spline <- predict(spline.fit, newdata = list(dis = dis.range),se = TRUE)
se.bands.spline <- cbind(dis.pred.spline$fit + 2*dis.pred.spline$se,
                  dis.pred.spline$fit - 2*dis.pred.spline$se)
plot(Boston$dis,Boston$nox,main = "Distance to employment centers vs NOX levels",
    ylab = "NOX levels", xlab = "weighted mean of distances to five Boston employment centers", col = "red")
lines(dis.range,dis.pred.spline$fit, col = "blue")
matlines(dis.range, se.bands.spline, col = "green",lwd = 2, lty = 2)
knots <- quantile(Boston$dis,ppoints(4))
abline( v = knots, col = "black", lwd = 0.5)
```

The knots were chosen to produce knots at uniform quantiles of the data. As observed, the splines produce a good fit for low distances but it also exhibits a tail problem.

**E.** Now fit a regression spline for a range of degrees of freedom, and
plot the resulting fits and report the resulting RSS. Describe the
results obtained.

```r
cv.error <- rep(NA,11)

for (d in 3:15){
    spline.fit <- glm(nox~bs(dis,df = d),data = Boston)
    distance.pred <- predict(spline.fit,newdata = Boston)
    RSS[d] <- mean((Boston$nox - distance.pred)^2)
}

plot(RSS,col = "red", lwd = 1 ,type = "b")
```

As expected, the residual sum of squares decreases monotonically as the number of degrees of freedom increases. This is expected as more degrees of freedom increase the flexibility of the model.
