**A.** Perform polynomial regression to predict wage using age. Use cross-validation to select the optimal degree d for the polynomial.
What degree was chosen, and how does this compare to the results of hypothesis testing using ANOVA? Make a plot of
the resulting polynomial fit to the data.

```r
library(boot)
library(ISLR)
```

```r
set.seed(200)
CV.error.10fold <- rep(0,5)
poly.model <- list()
degree = 1:5

for (d in degree){
  poly.model[[d]] <- glm(wage~poly(age,d),data = Wage)
  CV.error.10fold[d] <- cv.glm(data = Wage, glmfit = poly.model[[d]], K = 10)$delta[1]   
}

plot(degree,CV.error.10fold,type = 'b', col = "red",xlab = "Degree of polynomial", 
     ylab = "10 Fold Cross-Validation Error")
```

```r
which.min(CV.error.10fold)
```

```r
anova(poly.model[[1]],poly.model[[2]],poly.model[[3]],poly.model[[4]],poly.model[[5]],test = "F")
```

The cross-validation error is lowest with a 4 degree polynomial regression. On the other hand the ANOVA showed that the best model is cubic. Since there is no much difference in the error and the cubid model is simpler that is the model i would choose.

```r
age.pred <- seq(from = min(Wage$age),to = max(Wage$age))
best.poly.model <- glm(wage~poly(age,3),data = Wage)
poly.pred <- predict(best.poly.model,newdata = list(age = age.pred),se = TRUE)
se.bands <- cbind(poly.pred$fit + 2*poly.pred$se,poly.pred$fit - 2*poly.pred$se)
plot(Wage$age,Wage$wage,pch = 20,cex = 0.5,col = "darkgray")
lines(age.pred,poly.pred$fit,type = "l",col = "blue",lwd = 2)
matlines(age.pred, se.bands, col = "green",lwd = 2, lty = 2)
```

**B.** Fit a step function to predict wage using age, and perform crossvalidation
to choose the optimal number of cuts. Make a plot of
the fit obtained.

```r
CV.error.10fold.2 <- rep(NA,15)
step.model <- list()
ncuts = 2:15

for (n in ncuts){
    Wage$age.cuts <- cut(Wage$age,n)
    step.model <- glm(wage~age.cuts,data = Wage)
    CV.error.10fold.2[n] <- cv.glm(data = Wage, glmfit = step.model, K = 10)$delta[1]   
}

plot(ncuts,CV.error.10fold.2[-1],type = 'b', col = "red",xlab = "Number of cuts", 
     ylab = "10 Fold Cross-Validation Error")
```

Based on the graph, a low value for the CV Error is obtained at 8 cuts. This is the chosen model

```r
best.step.model <- glm(wage~cut(age,8),data = Wage)
step.pred <- predict(best.step.model,newdata = list(age = age.pred),se = TRUE)
step.se.bands <- cbind(step.pred$fit + 2*step.pred$se,step.pred$fit - 2*step.pred$se)
plot(Wage$age,Wage$wage,pch = 20,cex = 0.5,col = "darkgray")
lines(age.pred,step.pred$fit,type = "l",col = "blue",lwd = 2)
matlines(age.pred, step.se.bands, col = "green",lwd = 2, lty = 2)
```
