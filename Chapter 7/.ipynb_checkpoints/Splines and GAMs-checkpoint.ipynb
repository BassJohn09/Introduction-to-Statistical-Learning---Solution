## Splines

More flexible than polynomials and without the tail problem.

```r
require(ISLR)
require(splines)
```

```r
spline.fit <- lm(wage~ bs(age,knots = c(25,40,60)),data = Wage)
agelims <- range(Wage$age)
age.grid <- seq(from = agelims[1],to = agelims[2])
spline.preds <- predict(spline.fit,newdata = list(age = age.grid),se = TRUE)
spline.se.bands <- cbind(spline.preds$fit+2*spline.preds$se,spline.preds$fit-2*spline.preds$se)

plot(Wage$age,Wage$wage,col = "darkgray",pch = 20,cex = 0.5)
lines(age.grid,spline.preds$fit, col = "red", lwd = 2)
matlines(age.grid,spline.se.bands,col = "green", lwd = 2, lty = 2)
abline(v = c(25,40,60),col = "black",lwd = 0.5)
```

## Smoothing Splines 
***

```r
sspline.fit <- smooth.spline(Wage$age,Wage$wage,df = 16)

plot(Wage$age,Wage$wage,col = "darkgray",pch = 20,cex = 0.5)
lines(age.grid,spline.preds$fit, col = "red", lwd = 2)
abline(v = c(25,40,60),col = "black",lwd = 0.5)

lines(sspline.fit,col = "blue",lwd = 2)
```

Using LOOCV to determine the best parameter for the smoothing spline method

```r
sspline.fit.loocv <- smooth.spline(Wage$age,Wage$wage,cv = TRUE)
sspline.fit.loocv
```

```r
plot(Wage$age,Wage$wage,col = "darkgray",pch = 20,cex = 0.5)
lines(sspline.fit, col = "red", lwd = 2)
lines(sspline.fit.loocv,col = "purple",lwd = 2)
abline(v = c(25,40,60),col = "black",lwd = 0.5)
```

## Generalized Additive Models
***

```r
require(gam)
```

```r
gam1 <- gam(wage ~ s(age,df = 4)+s(year,df = 4)+education, data = Wage)
```

**NOTE:** s function is an special function within the gam package to do smooth splines to an specific variable

```r
par(mfrow = c(3,1))
plot(gam1,se = TRUE)
```

## Classification using GAMs

```r
gam2 <- gam(I(wage > 250) ~ s(age,df = 4)+year+education, data = Wage,family = "binomial")
plot(gam2)
```

```r
gam2a <- gam(I(wage > 250) ~ s(age,df = 4)+education, data = Wage,family = "binomial")
```

```r
anova(gam2a,gam2,test = "Chisq")
```

*Conclusion:* Variable year is not necessary in the classification model

```r
lm1 <- lm(wage ~ ns(age,df = 4)+ns(year,df = 4)+education,data = Wage)
plot.Gam(lm1,se = TRUE)
```

**Note:** ns is an special function in lm and gam to specify natural splines method
