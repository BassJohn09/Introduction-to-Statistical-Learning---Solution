**10.** This question relates to the College data set.

**A.** Split the data into a training set and a test set. Using out-of-state
tuition as the response and the other variables as the predictors,
perform forward stepwise selection on the training set in order
to identify a satisfactory model that uses just a subset of the
predictors.

```r
library (ISLR)
library(leaps)
library(gam)
```

```r
set.seed(100)
train.rows <- sample(1:nrow(College),630)
train.set <- College[train.rows,]
test.set <- College[-train.rows,]
fit.forward <- regsubsets(Outstate~.,data = train.set,nvmax= 17,method = "forward")
plot(fit.forward,scale = "bic")
```

The best model was chosen considering the simplest with a low BIC. A model containing 6 variables was chosen. 
the variables are: Private, Room.board, PhD, perc.alumni, Expend and Grad.Rate.


**B.** Fit a GAM on the training data, using out-of-state tuition as
the response and the features selected in the previous step as
the predictors. Plot the results, and explain your findings.

```r
cv.error <- rep(NA,10)

for (d in 2:11){
    sspline.fit.loocv <- smooth.spline(College$Room.Board,College$Outstate, df = d) 
    cv.error[d-1] <- sspline.fit.loocv$cv.crit
    }

plot(cv.error)
```

7 degrees of freedom are selected for the room board

```r
for (d in 2:11){
    sspline.fit.loocv <- smooth.spline(College$PhD,College$Outstate, df = d) 
    cv.error[d-1] <- sspline.fit.loocv$cv.crit
    }

plot(cv.error)
```

```r
3 degrees of freedom are selected for the PhD 
```

```r
for (d in 2:11){
    sspline.fit.loocv <- smooth.spline(College$perc.alumni,College$Outstate, df = d) 
    cv.error[d] <- sspline.fit.loocv$cv.crit
    }

plot(cv.error[2:11])
```

Linear model selected for the perc.alumni variable

```r
for (d in 2:11){
    sspline.fit.loocv <- smooth.spline(College$Expend,College$Outstate, df = d) 
    cv.error[d-1] <- sspline.fit.loocv$cv.crit
    }

plot(cv.error)
```

```r
4 degrees of freedom are selected for the Expend 
```

```r
for (d in 2:10){
    sspline.fit.loocv <- smooth.spline(College$Grad.Rate,College$Outstate, df = d) 
    cv.error[d-1] <- sspline.fit.loocv$cv.crit
    }

plot(cv.error[2:9])
```

```r
5 degrees of freedom are selected for the Grad.rate
```

```r
gam.model <- gam(Outstate~Private+s(Room.Board,df= 6)+s(PhD,df = 3)+perc.alumni+s(Expend,df = 3)+s(Grad.Rate,5),data = train.set)
par(mfrow = c(3,2))
plot(gam.model,se = TRUE)
```

As expected, The Out-of-state Tuition increases with the room and board costs, the type of university (public/private) and the number of Ph.D. faculty. Surprisingly as more alumni donates to the university, the Out-of-state tuition fee increases. On the other hand, there is a non-linear relationship between Expend and Out-of-state tuition.

```r
College[College$Grad.Rate > 100,]
```

**Note:** One data point of the graduation rate is above 100 and it is increasing the variance at the tail. Variance in Grad.rate diminishes significantly when removing this data point.

**C.** Evaluate the model obtained on the test set, and explain the
results obtained.

```r
gam.preds <- predict(gam.model, newdata = test.set)
gam.RSS <- mean((test.set$Outstate - gam.preds)^2)
Total <-mean((test.set$Outstate - mean(test.set$Outstate))^2) 
gam.R2 <- 1 - gam.RSS/Total
gam.R2
```

```r
linear.model <- lm(Outstate~Private+Room.Board+PhD+perc.alumni+Expend+Grad.Rate,data = train.set)
linear.preds <- predict(linear.model, newdata = test.set)
linear.RSS <- mean((test.set$Outstate - linear.preds)^2)
linear.R2 <- 1-linear.RSS/Total
linear.R2
```

Generalized additive model is only slighlty better at predicting than linear model at the cost of much higher complexity.
