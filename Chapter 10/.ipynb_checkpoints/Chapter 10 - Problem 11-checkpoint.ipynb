On the book website, www.StatLearning.com, there is a gene expression
data set (*Ch10Ex11.csv*) that consists of 40 tissue samples with
measurements on 1,000 genes. The first 20 samples are from healthy
patients, while the second 20 are from a diseased group.

**A.** Load in the data using read.csv(). You will need to select
header=F.

```r
data <- read.csv("Ch10Ex11.csv", head = FALSE)
head(data)
```

**B.** Apply hierarchical clustering to the samples using correlation-based
distance, and plot the dendrogram. Do the genes separate
the samples into the two groups? Do your results depend on the
type of linkage used?

```r
cor.data <- as.dist(1 - cor(scale(data)))
hc.complete <- hclust(cor.data,method = "complete")
plot(hc.complete)
```

```r
hc.single <- hclust(cor.data,method = "single")
plot(hc.single)
```

```r
hc.average <- hclust(cor.data,method = "average")
plot(hc.average)
```

The separation highly depends on the type of linkage used. When using complete linkage, the separation is very apparent. When using single linkage, the method tends to aggregate clusters one by one, therefore the separation occurs at a lower height and more clusters. The average cluster splits the data incorrectly. it assigns healthy people to the unhealthy cluster. That said, the separation is complete at a lower height.  

**C.** Your collaborator wants to know which genes differ the most
across the two groups. Suggest a way to answer this question,
and apply it here.

The question depends on what the difference in gene expression actually means. Assume we want to find out if the mean expression between the groups differs significantly, then a possible solution is to run a two-sample t test for each individual gene.   

```r
group <- cutree(hc.complete,4)
index.healthy <- which(group <= 3) #Ensure complete separation between healthy and unhealthy
index.diseased <- which(group == 4)

Healthy <- data[,index.healthy]
Diseased <- data[,index.diseased]
Healthy <- mapply(Healthy, FUN=as.numeric)
Diseased <- mapply(Diseased, FUN=as.numeric)
```

## Example for Gene 1

```r
gene.h <- Healthy[1,]
gene.d <- Diseased[1,]
hist(gene.h)
```

```r
hist(gene.d)
```

Check if samples are normally distributed

```r
shapiro.test(gene.h)
```

```r
shapiro.test(gene.d)
```

Both genes samples are normally distributed. Assuming the samples were taken independently then it is necessary to check homocedasticity

```r
var.test(gene.h,gene.d)
```

There is no enough evidence to reject null hypothesis, assuming equal variances.

```r
t.test(gene.h, gene.d, alternative = "two.sided", var.equal = TRUE)
```

There is no significant diference between the mean gene 1 expression of a healthy individual vs diseased group.

```r
gene.differ <- rep(NA,1000)
alpha <- 0.05

for (i in 1:1000){
    gene.h <- Healthy[i,]
    gene.d <- Diseased[i,]
    gene.h.normality <- shapiro.test(gene.h)
    gene.d.normality <- shapiro.test(gene.d)
    
    if (gene.h.normality$p.value > alpha & gene.d.normality$p.value > alpha) {
        var.equal.t <- var.test(gene.h,gene.d)
        var.equal <- ifelse(var.equal.t$p.value > alpha,TRUE,FALSE)
        two.sample.t.test <- t.test(gene.h, gene.d, alternative = "two.sided", var.equal = var.equal)
        
        if (two.sample.t.test$p.value < alpha) {
            gene.differ[i] = mean(gene.h) - mean(gene.d)
        }
    }
}

loc.diff <- which(gene.differ != is.na(gene.differ))
```

The following genes are expressed significantly different between the healthy and the diseased group. 

```r
loc.diff
```

```r
gene.differ <- na.omit(gene.differ)
gene.differ <- data.frame(loc.diff,gene.differ)
colnames(gene.differ) <- c("Gene","MeanDiff")

gene.differ[order(gene.differ$MeanDiff),]
```

Genes 127, 115, 81 are more expressed in the unhealthy people than healthy people. Genes 18, 21 and 36 are more expressed in healthy people.
