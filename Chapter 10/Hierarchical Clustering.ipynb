```r
set.seed(25000)
x <- matrix(rnorm(100^2),100,2)
xmeans <- matrix(rnorm(8,sd = 4), 4, 2) #displace gaussian distribution with random means with higher variance.
which <- sample(1:4,100,replace = TRUE)
x <- x + xmeans[which,] # Check the incantation formula used to distribute the means.
plot(x,col = which, pch = 19)
```

```r
hc.complete <- hclust(dist(x),method = "complete")
plot(hc.complete)
```

```r
hc.single <- hclust(dist(x),method = "single")
plot(hc.single)
```

```r
hc.average <- hclust(dist(x),method = "average")
plot(hc.average)
```

# Correlation-based distance 

It can be computed using the as.dist() function, which converts an arbitrary square symmetric matrix into a form that
the hclust() function recognizes as a distance matrix. However, this only
makes sense for data with at least three features since the absolute correlation
between any two observations with measurements on two features is
always 1

```r
x <- matrix(rnorm(30*3), ncol =3)
dd <- as.dist(1 - cor(t(x)))
plot(hclust(dd, method ="complete"), main=" Complete Linkage
with Correlation-Based Distance ", xlab="", sub ="")
```
