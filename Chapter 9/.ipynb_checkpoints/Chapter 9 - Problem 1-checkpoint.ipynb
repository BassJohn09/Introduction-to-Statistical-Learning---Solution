## Chapter 9 - Problem 1

```r
x1 <- seq(1,100)
plane1 <- function(x1){1 + 3*x1}
plane2 <- function(x2){(2-x1)/2}

x_pol1 <- c(0,0,100)
y_pol1 <- c(0,300,300)
x_pol2 <- c(0,100,100)
y_pol2 <- c(0,0,300)

plot(x1,plane1(x1),type = 'l',xlab = 'X1', ylab = 'X2',xlim = c(0,100), ylim = c(0,300))
polygon(x_pol1,y_pol1,col ="skyblue")
polygon(x_pol2,y_pol2,col ="orange1")
legend("topleft", legend = c("1 + 3*x1 - x2 < 0","1 + 3*x1 - x2 > 0"), col = c("skyblue","orange1"), lty = 19)
```

```r
x_pol1 <- c(0,0,100)
y_pol1 <- c(1,-49,-49)
x_pol2 <- c(0,100,100)
y_pol2 <- c(1,0,-49)

plot(x1,plane2(x1),type = 'l',xlab = 'X1', ylab = 'X2')
polygon(x_pol1,y_pol1,col ="skyblue")
polygon(x_pol2,y_pol2,col ="orange1")
legend("topleft", legend = c("−2 + x1 + 2*x2 < 0","−2 + x1 + 2*x2 > 0"), col = c("skyblue","orange1"), lty = 1)
```
