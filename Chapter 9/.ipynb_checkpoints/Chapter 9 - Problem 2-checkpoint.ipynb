**2.** We have seen that in p = 2 dimensions, a linear decision boundary
takes the form $β_0+β_1x_1+β_2x_2 = 0$.We now investigate a non-linear
decision boundary.

**A.** Sketch the curve
$$(1 + X_1)^2 + (2 − X_2)^2 = 4.$$


The curve is a circle of radius 2 and center (-1,2). To draw the circle correctly, check the limits of the equation.
* for x1. the limits are attained when $(2-x_2)^2 = 0$. the limits are $x_1 = -3$ and $x_1 = 1$. 
* for x2. the limits are attained when $(1+x_1)^2 = 0$. the limits are $x_2 = 0$ and $x_2 = 4$.

**Note:** all the limits are offset by one unit for better visualization

```r
plot(NA,NA,xlim = c(-4, 2), ylim = c(-1, 5),asp=1, xlab = "x1", ylab = "x2")
symbols(x=-1, y=2, circles=2, inches=F, add=T)
```

**B.** On your sketch, indicate the set of points for which
$$(1 + x_1)^2 + (2 − x_2)^2 > 4 $$ 
as well as the set of points for which
$$(1 + x_1)^2 + (2 − x_2)^2 ≤ 4 $$

```r
xgrid1 <- expand.grid(X1 = seq(from = -3,to = 1,length = 100), 
                      X2 = seq(from = 0,to = 2,length = 100)) #lattice function
ygrid1 <- rep("blue",n*n/2)
ygrid1[xgrid1$X2 < 2 - sqrt(4 - (1+xgrid1$X1)^2)] = "blue"
ygrid1[xgrid1$X2 >= 2 - sqrt(4 - (1+xgrid1$X1)^2)] = "red"

xgrid2 <- expand.grid(X1 = seq(from = -3,to = 1,length = 100), 
                      X2 = seq(from = 2,to = 4,length = 100)) #lattice function

ygrid2[xgrid2$X2 >= 2 + sqrt(4 - (1+xgrid2$X1)^2)] = "blue"
ygrid2[xgrid2$X2 < 2 + sqrt(4 - (1+xgrid2$X1)^2)] = "red"

plot(xgrid1, col = ygrid1, cex = 0.1, pch = 19, 
     type = 'p', ylim = c(0,4), xlab = "x1", ylab = "x2")
points(xgrid2,col = ygrid2, pch = 19, cex = 0.1)
```

**C** Suppose that a classifier assigns an observation to the blue class
if: $$(1 + x_1)^2 + (2 − x_2)^2 > 4$$ <br>
and to the red class otherwise. To what class is the observation
(0, 0) classified? (−1, 1)? (2, 2)? (3, 8)?

```r
x1 = c(0,-1,2,3)
x2 = c(0,1,2,8)

y = ifelse(x1 > -3 & x1 < 1 & 
           (x2 - 2 - sqrt(4 - (1+x1)^2))*(x2 - 2 + sqrt(4 - (1+x1)^2)) <= 0, "red","blue")


plot(NA,NA,xlim = c(-4, 2), ylim = c(-1, 8),asp=1, xlab = "x1", ylab = "x2", lwd = 0.5)
symbols(x=-1, y=2, circles=2, inches=F, add=T)
points(x1,x2,col = y, pch = 19)
```

**D.** Argue that while the decision boundary in (c) is not linear in terms of $x_1$ and $x_2$, 
it is linear in terms of $x_1$, $x_1^2$, $x_2$, $x_2^2$.

$$(1 + x_1)^2 + (2 − x_2)^2 = 4 $$ 
$$ x_1^2 + 2x_1 + 1 + x_2^2 - 4x_2 + 4 = 4 $$ 
$$ x_1^2 + 2x_1 + 1 + x_2^2 - 4x_2 = 0 $$ 

The decomposition of the decision boundary shows a linear equation in terms of $x_1$, $x_1^2$, $x_2$, $x_2^2$.
