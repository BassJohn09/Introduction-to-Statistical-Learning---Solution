# Chapter 9 - Problem 7

In this problem, you will use support vector approaches in order to
predict whether a given car gets high or low gas mileage based on the
Auto data set.

**A.** Create a binary variable that takes on a 1 for cars with gas
mileage above the median, and a 0 for cars with gas mileage
below the median.

```r
library(ISLR)
Highmpg <- ifelse(Auto$mpg > median(Auto$mpg),1,0)
```

**B.** Fit a support vector classifier to the data with various values
of cost, in order to predict whether a car gets high or low gas
mileage. Report the cross-validation errors associated with different
values of this parameter. Comment on your results.

```r
Auto$Highmpg <- as.factor(Highmpg)
```

```r
library(e1071)
cost <- c(0.001,0.01, 0.1, 1,10,100,1000,10000)
tune.out <- tune(svm, Highmpg ~ . - mpg, data = Auto, kernel = "linear", ranges = list(cost = cost))
summary(tune.out)
```

Cross validation error is minimized at 0.01

**C.** Now repeat (b), this time using SVMs with radial and polynomial
basis kernels, with different values of gamma and degree and
cost. Comment on your results.

```r
degrees <- c(2,3,4)
tune.out.poly <- tune(svm, Highmpg ~ . - mpg, data = Auto, kernel = "polynomial", 
                      ranges = list(cost = cost[2:5],degree = degrees))
summary(tune.out.poly)
```

The minimum cross validation error was attained at cost = 10 and degree = 2. It is worth noting that the error does not differ with different costs and degrees (except at the minimum).

```r
tune.out.radial <- tune(svm, Highmpg ~ . - mpg, data = Auto, kernel = "radial", 
                      ranges = list(cost = cost[2:5],gamma = cost))
summary(tune.out.radial)
```

The cross validation for the radial support vector machine is minimized at a $cost = 1$ and $\gamma = 1$.

*Note:* The minimum cross validation error of the polynomial SVM was significantly worse than the SVC and the radial SVM.


**D.** Make some plots to back up your assertions in (b) and (c).

```r
data = Auto[,2:ncol(Auto)]
svm.linear = svm(Highmpg ~ ., data = data, kernel = "linear", cost = 0.01)
svm.poly = svm(Highmpg ~ ., data = data, kernel = "polynomial", cost = 10, 
    degree = 2)
svm.radial = svm(Highmpg ~ ., data = data, kernel = "radial", cost = 1, gamma = 1)

#par(mfrow = c(2,1))
plot(svm.linear,data,weight~horsepower)
plot(svm.poly,data,weight~horsepower)
plot(svm.radial,data,weight~horsepower)
```

The high mileage of a car is highly correlated to low weight (below 3000 Kg)/ low horsepower (below (100 hp))

```r
plot(svm.linear,data,acceleration~weight)
plot(svm.poly,data,acceleration~weight)
plot(svm.radial,data,acceleration~weight)
```

High mileage is highly dependant on weight at low correlated to acceleration. Most of the high mileage data points are found at weight values below 3000 Kg. All SVM predict a similar behavior

```r
plot(svm.linear,data,cylinders~displacement)
plot(svm.poly,data,cylinders~displacement)
plot(svm.radial,data,cylinders~displacement)
```

High mileage is correlated to low number of cylinders and low displacement.

```r
plot(svm.linear,data,year~horsepower)
plot(svm.poly,data,year~horsepower)
plot(svm.radial,data,year~horsepower)
```

High mileage does not strongly correlate with the year of the car.
