**1.** This problem involves hyperplanes in two dimensions.

**A.** Sketch the hyperplane $1 + 3x_1 − x_2 = 0$. Indicate the set of
points for which $1 + 3x_1 − x_2 > 0$, as well as the set of points
for which $1 + 3x_1 − x_2 < 0$.

```r

plane1 <- function(x){1+3*x}
plane2 <- function(x){(2-x)/2}

n = 100
x1 <- seq(from = -10,to = 10, length = n)
x2 <- seq(from = min(min(plane1(x1)),min(plane2(x1))),to = max(max(plane1(x1)),max(plane2(x1))), length = 100)
xgrid1 <- expand.grid(X1 = x1, X2 = x2) #lattice function
ygrid <- rep("green",n*n)
ygrid[xgrid1$X2 > plane1(xgrid1$X1) & xgrid1$X2 < plane2(xgrid1$X1)] <- "red"
ygrid[xgrid1$X2 < plane1(xgrid1$X1) & xgrid1$X2 < plane2(xgrid1$X1)] <- "green"
ygrid[xgrid1$X2 > plane1(xgrid1$X1) & xgrid1$X2 > plane2(xgrid1$X1)] <- "blue"
ygrid[xgrid1$X2 < plane1(xgrid1$X1) & xgrid1$X2 > plane2(xgrid1$X1)] <- "orange"


plot(xgrid1, col = ygrid ,pch = 20, cex = .2)
#points(xgrid2, col = c("green","red")[ygrid2],pch = 20, cex = .2)
#legend("topleft", legend = c("1 + 3*x1 - x2 > 0","1 + 3*x1 - x2 < 0"), col = c("skyblue","orange1"), pch = 20)

```
