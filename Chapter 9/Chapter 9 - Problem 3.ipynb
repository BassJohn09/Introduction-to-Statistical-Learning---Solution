Here we explore the maximal margin classifier on a toy data set.

**A.** We are given n = 7 observations in p = 2 dimensions. For each
observation, there is an associated class label.

```r
x1 = c(3,2,4,1,2,4,4) 
x2 = c(4,2,4,4,1,3,1) 
y = c("red","red","red","red","blue","blue","blue")

plot(x1,x2,col = y, pch = 19)
```

**B.** Sketch the optimal separating hyperplane, and provide the equation for this hyperplane (of the form (9.1)).

```r
library(e1071) # Name of the function comes from a number of a room in the University of Berlin
dat <- data.frame(x1,x2,y=as.factor(y))
svmfit <- svm(y ~ .,data = dat, kernel = "linear", cost = 10, scale = FALSE) #Notice we specify the type of kernel and cost
print(svmfit)
```

```r
make.grid <- function (x, n = 75) {
    grange <- apply(x,2,range) #Range of each x value
    x1 <- seq(from = grange[1,1],to = grange[2,1], length = n)
    x2 <- seq(from = grange[1,2],to = grange[2,2], length = n)
    return(expand.grid(x1 = x1, x2 = x2)) #lattice function
}

xgrid <- make.grid(data.frame(x1,x2))
ygrid <- predict(svmfit,newdata = xgrid)
plot(xgrid, col = c("blue","red")[as.numeric(ygrid)],pch = 20, cex = .2)
points(x1, x2, col = y, pch = 19)

x.m <- matrix(c(x1,x2),7,2)
```

```r
beta = drop(t(svmfit$coefs)%*%x.m[svmfit$index,])
beta0 = svmfit$rho

beta
beta0
```

The equation is: $ -1 - 2x_1 + 2x_2 = 0 $

**D.** On your sketch, indicate the margin for the maximal margin hyperplane.

```r
plot(xgrid, col = c("blue","red")[as.numeric(ygrid)],pch = 20, cex = .2)
points(x1, x2, col = y, pch = 19)
abline(beta0/beta[2],-beta[1]/beta[2]) #the slope and intersection are derived from the formula
#Margin
abline((beta0-1)/beta[2],-beta[1]/beta[2], lty = 2)
abline((beta0+1)/beta[2],-beta[1]/beta[2], lty = 2)
```

**E.** Indicate the support vectors for the maximal margin classifier.

```r
plot(xgrid, col = c("blue","red")[as.numeric(ygrid)],pch = 20, cex = .2)
points(x1, x2, col = y, pch = 19)
abline(beta0/beta[2],-beta[1]/beta[2]) #the slope and intersection are derived from the formula
#Margin
abline((beta0-1)/beta[2],-beta[1]/beta[2], lty = 2)
abline((beta0+1)/beta[2],-beta[1]/beta[2], lty = 2)
points(x.m[svmfit$index,], pch = 5, cex = 2)
```
