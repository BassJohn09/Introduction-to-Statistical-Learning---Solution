**A.** Consider (6.12) with $p = 1$. For some choice of $y_1$ and λ > 0, 
plot (6.12) as a function of $β_1$. Your plot should confirm that (6.12) is solved by (6.14).

6.12 $$\sum_{j=1}^{p} (y_j-\beta_j)^2 + \lambda\sum_{j=1}^{p}\beta_j^2 $$

if $p = 1$, then 6.12 becomes:

$$ (1+\lambda)\beta_1^2-2y_1\beta_1+y_1^2$$

```r
EQ.6.12 <- function(b1,l,y1){
    fun <- (1+l)*b1^2-2*y1*b1+y1^2
    return(fun)
}

y1 <- 10
l <- 19

beta <- seq(-3,3,length = 50)
fun <- EQ.6.12(beta,l,y1)
plot(beta,fun,xlab = "Beta",ylab = "EQ 6.12",type = "b",pch = 19)

x.min = y1/(1+l)
y.min = EQ.6.12(x.min,l,y1)
points(x.min,y.min,col = "red",pch = 19)
```

**B.** Consider (6.13) with $p = 1$. For some choice of $y_1$ and λ > 0,
plot (6.13) as a function of $β_1$. Your plot should confirm that
(6.13) is solved by (6.15).

6.13. $$\sum_{j=1}^{p} (y_j-\beta_j)^2 + \lambda\sum_{j=1}^{p}|\beta_j| $$

if $p = 1$, then 6.13 becomes:

$$ \beta_1^2+\lambda|\beta_1|-2y_1\beta_1+y_1^2$$

```r
EQ.6.13 <- function(b1,l,y1){
    fun <- b1^2+l*abs(b1)-2*y1*b1+y1^2
    return(fun)
}

lasso.min <- function(y1,l){
    if (y1 > l/2){
        return(y1-l/2)
    } else if (y1 < -l/2){
        return(y1+l/2)
    } else {
        return(0)
    }
    
}

y1 <- 10
l <- 20

beta <- seq(-3,3,length = 50)
fun <- EQ.6.13(beta,l,y1)
plot(beta,fun,xlab = "Beta",ylab = "EQ 6.13",type = "b",pch = 19)

x.min = lasso.min(y1,l)
y.min = EQ.6.13(x.min,l,y1)
points(x.min,y.min,col = "red",pch = 19)
```
