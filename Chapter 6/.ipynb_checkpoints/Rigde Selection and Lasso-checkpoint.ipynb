## Rigde Selection and Lasso


**glmnet library** will be used, this library does not use the formula methods used typically in R. 
Instead x and y datasets have to be provided.

```r
library(glmnet)
library(ISLR)
Hitters <- na.omit(Hitters)

x <- model.matrix(Salary~.-1,data = Hitters) #Eliminates intercept
y <- Hitters$Salary
```

```r
summary(x)
```

glmnet function uses parameter alpha to define which model is used: <br>
alpha = 0 <- Rigde regression  <br>
alpha = 1 <- Lasso  <br>
0 < alpha < 1 <- Elastic models (models between ridge and lasso)

```r
param <- seq(5,-2,length = 100)
lambdas <- 10^param
fit.ridge <- glmnet(x,y,alpha = 0, family = "gaussian",lambda = lambdas)
plot(fit.ridge,xvar = "lambda",label = TRUE)
```

Cross validation for ridge regression and Lasso
***

```r
cv.ridge <- cv.glmnet(x,y,alpha = 0)
plot(cv.ridge)
```

## Lasso Model

```r
fit.lasso <- glmnet(x,y,alpha = 1)
plot(fit.lasso,xvar = "dev",label = TRUE)
```

```r
cv.lasso <- cv.glmnet(x,y,alpha = 1)
plot(cv.lasso)
```

```r
coef(cv.lasso) #Select best model automatically
```

```r
train <- sample(seq(nrow(Hitters)),180,replace = FALSE)
fit.lasso.tr <- glmnet(x[train,],y[train],alpha = 1,nlambda = 100)
fit.lasso.tr

pred <- predict(fit.lasso.tr,x[-train,])
dim(pred)
```

```r
rmse <- sqrt(apply((y[-train]-pred)^2,2,mean)) 
```

Notice that **y[test]** and pred have different dimensions but they can be used in the function.
This occurs because the vector **y[test]** is reused over the whole range of lambdas. 
It is a useful trick in R

```r
plot(log(fit.lasso.tr$lambda),rmse,type = "b",xlab = "log(Lambda)")
```

```r
lasso.best <- fit.lasso.tr$lambda[order(rmse)][1]
lasso.best
coef(fit.lasso.tr,s = lasso.best)
```
