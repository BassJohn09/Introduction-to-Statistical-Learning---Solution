3. Consider the Gini index, classification error, and cross-entropy in a
simple classification setting with two classes. Create a single plot
that displays each of these quantities as a function of $\hat{p}_{m1}$. The xaxis
should display $\hat{p}_{m1}$, ranging from 0 to 1, and the y-axis should
display the value of the Gini index, classification error, and entropy.

```r
P <- seq(0,1,0.01)

classifier <- function(p){
    return(1-max(c(p,(1-p))))
}

class.error <- lapply(P,classifier) #Classification Error
Gini.error <- P*(1-P)+(1-P)*P # Gini Index
Entropy.error <- -P*log(P)-(1-P)*log(1-P) # Cross-Entropy
    
Results = cbind(class.error,Gini.error,Entropy.error)

matplot(P,Results,type = "b", xlab = "Probability of Class 1", ylab = "Error", 
        col = c ("green","black","red"), pch = 19)
legend("topright",c("Classification Error", "Gini Index", "Cross-Entropy"), 
       col = c ("green","black","red"),pch = 19)
```
