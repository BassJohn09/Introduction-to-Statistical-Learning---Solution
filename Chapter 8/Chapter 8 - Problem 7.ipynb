7. In the lab, we applied random forests to the Boston data using mtry=6
and using ntree=25 and ntree=500. Create a plot displaying the test
error resulting from random forests on this data set for a more comprehensive
range of values for mtry and ntree. You can model your
plot after Figure 8.10. Describe the results obtained.

```r
require(randomForest)
require(MASS)
set.seed(200)
train <- sample(1:nrow(Boston),300)

```

```r
Test.error <- matrix(NA,nrow = 500,ncol = 4)
nvar <- c(12,8,4,5) #p, p/2, sqrt(p)
x.train <- Boston[train,-14]
y.train <- Boston[train,14]
x.test <- Boston[-train,-14]
y.test <- Boston[-train,14]

for (i in 1:4){
    Boston.RF <- randomForest(x = x.train, y = y.train, xtest = x.test, ytest = y.test,
                    mtry = nvar[i], ntree = 500)
    Test.error[,i] = Boston.RF$test$mse  
    }
```

```r
matplot(1:500,Test.error, pch = 20, type = "l", lwd = 0.3, cex = 0.5, 
        col = c("red","black","blue","green"), xlab = "Number of Trees", ylab = "Test Error")
legend("topright", c("m = p","m = p/2","m = sqrt(p)","m = 5")
       , col = c("red","black","blue","green"), pch = 20)
```

1. At low number of trees, the test error is considerably high regardless of the number of variables.
2. The test error is stable above 100 trees for all number of variables.
3. The lowest stable error was attained with m = p /2.
4. There are local minima for all cases below 100 trees. However these minima are a product of the data used for the test error not the method itself, conclusions around these datapoints might be faulty. 


```r
oob.err <- matrix(NA,nrow = 13, ncol = 5)
test.err <- matrix(NA,nrow = 13, ncol = 5)
ntrees <- c(10,50,100,200,500)

for (nvar in 1:13){
    for (i in 1:5){   
    fit <- randomForest(medv~.,data = Boston, subset = train, mtry = nvar, ntree = ntrees[i])
    oob.err[nvar,i] <- fit$mse[length(fit$mse)]
    pred <- predict(fit,newdata = Boston[-train,])
    test.err[nvar,i] <- with(Boston[-train,],mean((medv-pred)^2)) 
    }
}
```

```r
matplot(1:13, test.err, pch = 20, lwd = 0.5, type = "b",
       ylab = "Test error",xlab = "Number of variables per split", 
        col = c("black","red","green","blue", "purple"))
legend("topright", c("ntree = 10","ntree = 50","ntree = 100","ntree = 200","ntree = 500")
       , col = c("black","red","green","blue", "purple"), pch = 20)
```

1. The worst test error was obtained with the lowest number of trees (10).
2. The test error shows a smoother behaviour with increasing number of trees. 
3. The number of variables per split depends on the number of trees. at 100, the minimum error is about around 7. at 500, the minimum error is with 8 variables per split. 
when performing cross validation, consider both variables to make a reasonable conclusion.
